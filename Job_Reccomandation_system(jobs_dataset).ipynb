```python
import numpy as np
import pandas as pd
```

```python
from google.colab import files
path=list(files.upload().keys())[0]

```

```python
df=pd.read_csv(path)
df.head()
```

```python
df=df.sample(5000,ignore_index=True)
```

```python
import string
```

```python
panch=string.punctuation
panch
```

```python
def remove_panch(text):
  mylist=[]
  for i in text.split():
    if i not in panch:
      mylist.append(i)
  return " ".join(mylist)
```

```python
df['Job Title']=df['Job Title'].apply(remove_panch)
```

```python
df['Role Category']=df['Role Category'].apply(remove_panch)
```

```python
df['Industry']=df['Industry'].apply(remove_panch)
```

```python
df['Job Salary']=df['Job Salary'].apply(remove_panch)
```

```python
df['Content']=df.apply(lambda x: " ".join(x.dropna().astype(str)),axis=1)
```

```python
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000)
metrics=cv.fit_transform(df['Content']).toarray()
```

```python

metrics.shape
```

```python
similarity=cosine_similarity(metrics)
```

```python
def recommend(Job_Title):
  index=df[df['Job Title']==Job_Title].index[0]
  distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
  new_index=[i[0] for i in distances[1:11]]
  return df.iloc[new_index,:-1]
```

```python
df.sample(5)
```

```python
recommend('Software Developer')
```
